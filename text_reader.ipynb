```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("img2.png")
```

```python
output = reader.readtext('img2.png')
```

```python
output
```

```python
result=""
for i in range(len(output)):
  result+=output[i][1]+" "
result
```

```python
def read_text(img):
  output = reader.readtext(img)
  result=""
  for i in range(len(output)):
    result+=output[i][1]+" "
  return result
```

```python
read_text("img1.png")
```

```python

vidcap = cv2.VideoCapture("video.mp4")
strings=[]
count = 0 
success=True
while success:
    success, image = vidcap.read()

    # s="name"+str(count)+".png"
    if count%20==0:
      cv2.imwrite("a"+str(count)+".jpg",image )
      strings.append("a"+str(count)+".jpg")
    count+=1
```

```python
res=""
for i in strings:
  res+=read_text(i)+" "
res
```
